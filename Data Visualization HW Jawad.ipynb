```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
data_url = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv"
data_desc_url = "https://github.com/owid/covid-19-data/blob/master/public/data/owid-covid-codebook.csv"

#DataFrame of the covid data
df = pd.read_csv(data_url)
```

# Line plot for total deaths

Why I chose this: <br>
I chose this visualization for 5 of the western-european countries to hint how good their healthcare service were. For any country, if we see a steep rise in total death, it could either mean there were not adequate healthcare or there were a lot of vulnerable population. <br>

Interpretation: <br>
As we can see from the graph, all contries had similar slopes (more accurately, if we took the first derivative of the curve, we would get more or less similar values) except Netherlands. It means Netherland handled the Covid-situation better than the other 4 countries

```python
#analyzing 3 years 6 months of data
start_date = '2020-03-01'
end_date = '2023-09-01'

# filtering dataframes
# usa_data = df[(df['iso_code'] == 'USA') & (df['date'] >= start_date) & (df['date'] <= end_date)] 
uk_data = df[(df['iso_code'] == 'GBR') & (df['date'] >= start_date) & (df['date'] <= end_date)]
deu_data = df[(df['iso_code'] == 'DEU') & (df['date'] >= start_date) & (df['date'] <= end_date)]
fra_data = df[(df['iso_code'] == 'FRA') & (df['date'] >= start_date) & (df['date'] <= end_date)]
ned_data = df[(df['iso_code'] == 'NLD') & (df['date'] >= start_date) & (df['date'] <= end_date)]
ita_data = df[(df['iso_code'] == 'ITA') & (df['date'] >= start_date) & (df['date'] <= end_date)]

legend_list = ["UK", "GERMANY", "FRANCE", "NETHERLAND", "ITALY"]
# Plotting
plt.figure(figsize=(15, 6))
# plt.plot(usa_data['date'], usa_data['total_deaths'], linestyle='-', color= 'blue')
plt.plot(uk_data['date'], uk_data['total_deaths'], linestyle='-', color= 'red')
plt.plot(deu_data['date'], deu_data['total_deaths'], linestyle='-', color= 'green')
plt.plot(fra_data['date'], fra_data['total_deaths'], linestyle='-', color= 'blue')
plt.plot(ned_data['date'], ned_data['total_deaths'], linestyle='-', color= 'purple')
plt.plot(ita_data['date'], ita_data['total_deaths'], linestyle='-', color= 'cyan')

title_font = {'fontsize': 16, 'fontweight': 'bold', 'fontfamily': 'Times New Roman'}
xlabel_font = {'fontsize': 14, 'fontweight': 'bold', 'fontfamily': 'Times New Roman'}
ylabel_font = {'fontsize': 14, 'fontweight': 'bold', 'fontfamily': 'Times New Roman'}

# Adding labels and title
plt.title('Line plot of total deaths in 5 West-European countries ({} to {})'.format(start_date, end_date), fontdict=title_font)
plt.xlabel('Date (2M 15D interval)', fontdict=xlabel_font)
plt.ylabel('Total Deaths', fontdict=ylabel_font)
plt.legend(legend_list)

# Rotating x-axis and skipping labels for better visibility
plt.xticks(uk_data['date'][::75], rotation=60)

plt.show()
```

# Line plot for total cases

Why I chose this: <br>
I chose this visualization for 5 of the western-european countries to get an idea of how well contained the pandemic was. For any country, if we see a steep rise in total cases, it means the country performed poorly to contain the situatuion in that particular time window. <br>

Interpretation: <br>
As we can see from the graph, France, Germany & Italy has almost similar slopes (more accurately, if we took the first derivative of the curve, we would get more or less similar values). Netherlands and UK also has similar slope which is different than the former. It means that compared to UK and Netherlands, the rest of the 3 countries performed poorly to contain the virus

```python
#analyzing 3 years 6 months of data
start_date = '2020-03-01'
end_date = '2023-09-01'

# filtering dataframes
# usa_data = df[(df['iso_code'] == 'USA') & (df['date'] >= start_date) & (df['date'] <= end_date)] 
uk_data = df[(df['iso_code'] == 'GBR') & (df['date'] >= start_date) & (df['date'] <= end_date)]
deu_data = df[(df['iso_code'] == 'DEU') & (df['date'] >= start_date) & (df['date'] <= end_date)]
fra_data = df[(df['iso_code'] == 'FRA') & (df['date'] >= start_date) & (df['date'] <= end_date)]
ned_data = df[(df['iso_code'] == 'NLD') & (df['date'] >= start_date) & (df['date'] <= end_date)]
ita_data = df[(df['iso_code'] == 'ITA') & (df['date'] >= start_date) & (df['date'] <= end_date)]

legend_list = ["UK", "GERMANY", "FRANCE", "NETHERLAND", "ITALY"]
# Plotting
plt.figure(figsize=(15, 6))
# plt.plot(usa_data['date'], usa_data['total_deaths'], linestyle='-', color= 'blue')
plt.plot(uk_data['date'], uk_data['total_cases'], linestyle='-', color= 'red')
plt.plot(deu_data['date'], deu_data['total_cases'], linestyle='-', color= 'green')
plt.plot(fra_data['date'], fra_data['total_cases'], linestyle='-', color= 'blue')
plt.plot(ned_data['date'], ned_data['total_cases'], linestyle='-', color= 'purple')
plt.plot(ita_data['date'], ita_data['total_cases'], linestyle='-', color= 'cyan')

title_font = {'fontsize': 16, 'fontweight': 'bold', 'fontfamily': 'Times New Roman'}
xlabel_font = {'fontsize': 14, 'fontweight': 'bold', 'fontfamily': 'Times New Roman'}
ylabel_font = {'fontsize': 14, 'fontweight': 'bold', 'fontfamily': 'Times New Roman'}

# Adding labels and title
plt.title('Line plot of total cases in 5 West-European countries ({} to {})'.format(start_date, end_date), fontdict=title_font)
plt.xlabel('Date (2M 15D interval)', fontdict=xlabel_font)
plt.ylabel('Total Cases', fontdict=ylabel_font)
plt.legend(legend_list)

# Rotating x-axis and skipping labels for better visibility
plt.xticks(uk_data['date'][::75], rotation=60)

plt.show()
```

# Box plot for positive rate

Why I chose this: <br>
Positive rate means the percentage of people tested positive after taking a test. The box plot neatly gives us an idea about the distribution.  <br>

Interpretation: <br>
From the graph we can interpret that Germany, Netherland and UK has a lot of outliers (data point outside the whiskers). But France and Italy does not have so many outliers. It means we can model the distribution for positive rate pretty much accurately for the France and Italy. For the rest of the countries, the distribuition will be heavily skewed. Also, it is notable that France, Germany, Italy and UK has almost same median.

```python
countries = ['GBR', 'DEU', 'FRA', 'NLD', 'ITA']
pr_and_country = df[df['iso_code'].isin(countries)][['positive_rate', 'iso_code']]
pr_and_country = pr_and_country.dropna()
# print(pr_and_country)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='iso_code', y='positive_rate', data=pr_and_country)

# Set plot labels and title
plt.title('Boxplots of Positive Rate for the five aformentioned Countries', fontdict=title_font)
plt.xlabel('Country', fontdict=xlabel_font)
plt.ylabel('Positive Rate', fontdict=ylabel_font)

# Show the plot
plt.show()
```

# Scatter plot : Reproduction Rate vs Total Vaccination per Hundred

Why I chose this: <br>
I wanted to find a correlation between the reproduction rate and vaccination. My initial hypothesis was as more people were vaccinated, the reproduction rate of the virus would slowly decrease, resulting in a negative correlation  <br>

Interpretation: <br>
From the graph we can see that all the countries follow almost the same trend. My initial hypothesis was proven wrong, there is no negative correlation. In fact, it looks very chaotic. Therefore I believe getting vaccinated does not greatly affect the reproduction rate of the virus 

```python
rr_and_vacc = df[df['iso_code'].isin(countries)][['reproduction_rate', 'total_vaccinations_per_hundred', 'iso_code']]

rr_and_vacc = rr_and_vacc.dropna()

# print(rr_and_vacc)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_vaccinations_per_hundred', y='reproduction_rate', hue='iso_code', data=rr_and_vacc, palette='Set1')

# Set plot labels and title
plt.title('Scatter Plots of Total Vaccinations per Hundred vs Reproduction Rate', fontdict=title_font)
plt.xlabel('Total Vaccinations per hundred', fontdict=xlabel_font)
plt.ylabel('Reproduction Rate', fontdict= ylabel_font)

# Show the legend
plt.legend(title='Country')

# Show the plot
plt.show()
```
